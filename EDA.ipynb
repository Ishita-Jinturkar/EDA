```python
import pandas as pd

url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Liver%20Data.csv"

df = pd.read_csv(url)
print(df.head())
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='all' ,axis=1)
```

```python
print(df.columns)
df.columns = df.columns.str.strip()
```

```python

# Handling missing values by replacing with column means
age_mean = df['Age of the patient'].mean()
df['Age of the patient'].fillna(age_mean, inplace=True)

total_bilirubin_mean = df['Total Bilirubin'].mean()
df['Total Bilirubin'].fillna(total_bilirubin_mean, inplace=True)

direct_bilirubin_mean = df['Direct Bilirubin'].mean()
df['Direct Bilirubin'].fillna(direct_bilirubin_mean, inplace=True)

alkphos_mean = df['Alkphos Alkaline Phosphotase'].mean()
df['Alkphos Alkaline Phosphotase'].fillna(alkphos_mean, inplace=True)

sgpt_mean = df['Sgpt Alamine Aminotransferase'].mean()
df['Sgpt Alamine Aminotransferase'].fillna(sgpt_mean, inplace=True)

sgot_mean = df['Sgot Aspartate Aminotransferase'].mean()
df['Sgot Aspartate Aminotransferase'].fillna(sgot_mean, inplace=True)

total_proteins_mean = df['Total Protiens'].mean()
df['Total Protiens'].fillna(total_proteins_mean, inplace=True)

albumin_mean = df['ALB Albumin'].mean()
df['ALB Albumin'].fillna(albumin_mean, inplace=True)

ag_ratio_mean = df['A/G Ratio Albumin and Globulin Ratio'].mean()
df['A/G Ratio Albumin and Globulin Ratio'].fillna(ag_ratio_mean, inplace=True)

```

```python
import numpy as np
df[np.abs(df['Alkphos Alkaline Phosphotase'])>700]
df[np.abs(df['Sgot Aspartate Aminotransferase'])>75]
```

```python

import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")
```

Univariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Bilirubin'], kde=True, color='skyblue')
plt.title('Distribution of Total Bilirubin')
plt.show()

# 2. Boxplot for Total Bilirubin
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Total Bilirubin'], color='lightgreen')
plt.title('Boxplot of Total Bilirubin')
plt.show()

# 3. Violin Plot for Total Bilirubin
plt.figure(figsize=(10, 6))
sns.violinplot(y=df['Total Bilirubin'], color='lightcoral')
plt.title('Violin Plot of Total Bilirubin')
plt.show()
```

# Bivariate Analysis

```python

# 4. Scatterplot between Age of the patient and Total Bilirubin
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age of the patient'], y=df['Total Bilirubin'], hue=df['Gender of the patient'], palette='cool')
plt.title('Scatterplot: Age vs Total Bilirubin')
plt.show()

# 5. Violin Plot between Gender and Total Bilirubin
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Gender of the patient'], y=df['Total Bilirubin'], palette='muted')
plt.title('Violin Plot: Gender vs Total Bilirubin')
plt.show()

# 6. Swarm Plot for Gender and Total Bilirubin
plt.figure(figsize=(10, 6))
sns.swarmplot(x=df['Gender of the patient'], y=df['Total Bilirubin'], color='purple')
plt.title('Swarm Plot: Gender vs Total Bilirubin')
plt.show()

# 7. Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Multivariate Analysis

```python

# 8. Pairplot (Numerical Variables Only)
plt.figure(figsize=(12, 8))
sns.pairplot(df, diag_kind="kde", hue="Gender of the patient", markers=["o", "s"])
plt.show()

# 9. 3D Scatterplot (Age, Total Bilirubin, Alkphos Alkaline Phosphotase)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Age of the patient'], df['Total Bilirubin'], df['Alkphos Alkaline Phosphotase'],
           c=np.random.rand(len(df)), cmap='plasma')

ax.set_xlabel('Age of the patient')
ax.set_ylabel('Total Bilirubin')
ax.set_zlabel('Alkphos Alkaline Phosphotase')
plt.title('3D Scatterplot: Age, Total Bilirubin, Alkphos Alkaline Phosphotase')
plt.show()
```
