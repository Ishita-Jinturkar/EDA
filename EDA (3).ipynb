```python

# Handling missing values by replacing with column means
age_mean = df['Age of the patient'].mean()
df['Age of the patient'].fillna(age_mean, inplace=True)

total_bilirubin_mean = df['Total Bilirubin'].mean()
df['Total Bilirubin'].fillna(total_bilirubin_mean, inplace=True)

direct_bilirubin_mean = df['Direct Bilirubin'].mean()
df['Direct Bilirubin'].fillna(direct_bilirubin_mean, inplace=True)

alkphos_mean = df['Alkphos Alkaline Phosphotase'].mean()
df['Alkphos Alkaline Phosphotase'].fillna(alkphos_mean, inplace=True)

sgpt_mean = df['Sgpt Alamine Aminotransferase'].mean()
df['Sgpt Alamine Aminotransferase'].fillna(sgpt_mean, inplace=True)

sgot_mean = df['Sgot Aspartate Aminotransferase'].mean()
df['Sgot Aspartate Aminotransferase'].fillna(sgot_mean, inplace=True)

total_proteins_mean = df['Total Protiens'].mean()
df['Total Protiens'].fillna(total_proteins_mean, inplace=True)

albumin_mean = df['ALB Albumin'].mean()
df['ALB Albumin'].fillna(albumin_mean, inplace=True)

ag_ratio_mean = df['A/G Ratio Albumin and Globulin Ratio'].mean()
df['A/G Ratio Albumin and Globulin Ratio'].fillna(ag_ratio_mean, inplace=True)

```

```python

df.columns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
X = df.iloc[:, [3, 4]].values
wcss = []
for i in range(1, 11):
 kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
 kmeans.fit(X)
 wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red',
label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue',
label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100,
c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan',
label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100,
c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,
1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
X = df.iloc[:, [3, 4]].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='purple', s=50, label='PCA projection')
plt.title('PCA of Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[y_kmeans == 0, 0], X_pca[y_kmeans == 0, 1], s=100,
c='red', label='Cluster 1')
plt.scatter(X_pca[y_kmeans == 1, 0], X_pca[y_kmeans == 1, 1], s=100,
c='blue', label='Cluster 2')
plt.scatter(X_pca[y_kmeans == 2, 0], X_pca[y_kmeans == 2, 1], s=100,
c='green', label='Cluster 3')
plt.scatter(X_pca[y_kmeans == 3, 0], X_pca[y_kmeans == 3, 1], s=100,
c='cyan', label='Cluster 4')
plt.scatter(X_pca[y_kmeans == 4, 0], X_pca[y_kmeans == 4, 1], s=100,
c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,
1], s=300, c='yellow', label='Centroids')
plt.title('K-Means Clusters in PCA Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
# Explained Variance ratio (to understand how much variance is explained by the 2 principal components)
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

numeric_cols = df.select_dtypes(include=['number'])
print("Numeric Columns in Dataset:")
print(numeric_cols.columns)

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="RdYlGn", square=True)
plt.title("Correlation Heatmap")
plt.show()
sns.lmplot(x='Total Bilirubin', y='Result', data=df, height=6, aspect=1.5)
plt.title("Simple Linear Regression: Total Bilirubin vs Result")
plt.show()

X_simple = df[['Total Bilirubin']]
y_simple = df[['Result']]


X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y_simple, test_size=0.3, random_state=10)

regressor_simple = LinearRegression()
regressor_simple.fit(X_train_simple, y_train_simple)

y_pred_simple = regressor_simple.predict(X_test_simple)


print("\nSimple Linear Regression Evaluation:")
print(f"R2 score: {r2_score(y_test_simple, y_pred_simple)}")
print(f"MAE score: {mean_absolute_error(y_test_simple, y_pred_simple)}")
print(f"MSE score: {mean_squared_error(y_test_simple, y_pred_simple)}")

evaluate_simple = pd.DataFrame({'Actual': y_test_simple.values.flatten(), 'Predicted': y_pred_simple.flatten()})
print("\nComparison (Simple Linear Regression):")
print(evaluate_simple.head(10))
evaluate_simple.head(10).plot(kind='bar', figsize=(8, 5))
plt.title("Actual vs Predicted (Simple Linear Regression)")
plt.show()

X_multi = df[['Total Bilirubin', 'Direct Bilirubin', 'ALB Albumin']]
y_multi = df[['Result']]


X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.3, random_state=10)

regressor_multi = LinearRegression()
regressor_multi.fit(X_train_multi, y_train_multi)

y_pred_multi = regressor_multi.predict(X_test_multi)


print("\nMultiple Linear Regression Evaluation:")
print(f"R2 score: {r2_score(y_test_multi, y_pred_multi)}")
print(f"MAE score: {mean_absolute_error(y_test_multi, y_pred_multi)}")
print(f"MSE score: {mean_squared_error(y_test_multi, y_pred_multi)}")

evaluate_multi = pd.DataFrame({'Actual': y_test_multi.values.flatten(), 'Predicted': y_pred_multi.flatten()})
print("\nComparison (Multiple Linear Regression):")
print(evaluate_multi.head(10))
evaluate_multi.head(10).plot(kind='bar', figsize=(8, 5))
plt.title("Actual vs Predicted (Multiple Linear Regression)")
plt.show()
```
